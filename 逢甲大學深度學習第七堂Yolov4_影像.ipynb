yolov3_custom.cfg 設定檔內容說明

classes ==>類別數量

batch ==> 每次抓取的圖片數量進行訓練

subdivisions ==> 每次要拆幾組(建議: 12G GPU => batch=24
subdivisions=8)

filter大小:(c+5)B

c ==> 類別的數量

B ==> 每一個Feature Map 可以偵測的bndBox數兩,預設為3

通常一般會設定3個，因為RGB三原色
